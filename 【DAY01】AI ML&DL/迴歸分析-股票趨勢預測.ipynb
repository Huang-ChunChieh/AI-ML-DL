# 資料前處理-分割訓練與測試資料

分離測試與訓練資料

我們設定資料集中前面70%的資料為訓練資料

後面30%為測試/驗證資料

# 定義迴歸的 y 與 x 

<p>資料先進行轉置</p>
<p> .reshape(變成幾個row , 變成幾個col)</p>
<p> 如果輸入 -1 => 資料原本長度 </p>

```python
# 常用的統計，向量分面的數學運算

import numpy as np
```

設定收盤價格為因變數

設定時間為自變數

在此的.reshape(-1,1)是把橫的array變成數列

```python
# 訓練資料
train_y = train_data['close']
train_x = np.arange(train_data.shape[0]).reshape(-1,1) 

# 測試資料
test_y = test_data['close']
test_x = np.arange(train_data.shape[0], stock_data.shape[0]).reshape(-1,1)
```

顯示訓練資料的因變數(y)的資料形態

```python
print('資料型態： ', type(train_y))
```

### Series 資料型態可以直接做統計運算

```python
print("std  : ", train_y.std())   ## 標準差
print("median  : ", train_y.median())   ## 中位數
print("mean : ", train_y.mean())  ## 平均值
print("sum  : ", train_y.sum())   ## 全部加總
```

# 線性迴歸模型建立

引進Scikit Learn機器學習套件 中的線性迴歸

設定一個空的模型放入lr, 等一下要往模型放東西

```python
from sklearn.linear_model import LinearRegression

lr =  LinearRegression() 
lr.fit(train_x, train_y) ### 迴歸訓練
```

叫出線性迴歸模型的結果

```python
print("b 斜率：", lr.coef_)
print("a 截距：", lr.intercept_)
print("迴歸公式： y = %.4f + %.4f * x "%( lr.intercept_, lr.coef_  ))
```

### 趨勢線

利用上面訓練出的線性迴歸結果來畫圖

就是 y = ax+b

```python
train_trend_line = lr.intercept_ + lr.coef_ * train_x
```

```python
# 畫出有趨勢線的K線圖
plotStockCandle(train_data, mav=(), subdata_list=[train_trend_line])
```

用.std（）叫出標準差

因爲五線譜中除了中間的回歸線以外,就是正負標準差或2個標準差

```python
train_lr_line_add_1_std = train_trend_line + train_y.std()
train_lr_line_sub_1_std = train_trend_line - train_y.std()
train_lr_line_add_2_std = train_trend_line + train_y.std() * 2
train_lr_line_sub_2_std = train_trend_line - train_y.std() * 2
# 畫出訓練期間的五線譜K線圖
plotStockCandle(
    train_data,
    mav=(),
    subdata_list=[
        train_trend_line,
        train_lr_line_add_1_std,
        train_lr_line_sub_1_std,
        train_lr_line_add_2_std,
        train_lr_line_sub_2_std,
    ])
```

# 拿測試資料來驗證一下

```python
test_trend_line = lr.intercept_ + test_x * lr.coef_ 
test_lr_line_add_1_std = test_trend_line + train_y.std()
test_lr_line_sub_1_std = test_trend_line - train_y.std() 
test_lr_line_add_2_std = test_trend_line + train_y.std() * 2
test_lr_line_sub_2_std = test_trend_line - train_y.std() * 2
# 畫出測試期間的五線譜K線圖
plotStockCandle(
    test_data, 
    mav=(),
    subdata_list=[
        test_trend_line,
        test_lr_line_add_1_std,
        test_lr_line_sub_1_std,
        test_lr_line_add_2_std,
        test_lr_line_sub_2_std,
    ])
```

# 把所有的資料放進模型迴歸後的五線譜

```python
all_x = np.arange(stock_data.shape[0])
```

```python
all_trend_line = lr.intercept_ + all_x * lr.coef_
all_lr_line_add_1_std = all_trend_line + train_y.std()
all_lr_line_sub_1_std = all_trend_line - train_y.std()
all_lr_line_add_2_std = all_trend_line + train_y.std() * 2
all_lr_line_sub_2_std = all_trend_line - train_y.std() * 2

plotStockCandle(
    stock_data, 
    mav=(),
    vlines_date = last_train_date,
    subdata_list=[
        all_trend_line,
        all_lr_line_add_1_std,
        all_lr_line_sub_1_std,
        all_lr_line_add_2_std,
        all_lr_line_sub_2_std,
    ])
```
